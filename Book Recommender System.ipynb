```python
def recommend(book_name):
    index = np.where(pt.index ==book_name)[0][0]
    similar_items  = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
    
```

```python
import pickle
from sklearn.metrics.pairwise import cosine_similarity

# Ensure required data is already processed
# Merge & Clean
data = books.merge(ratings, on='ISBN')
data.dropna(inplace=True)

# Popular books
num_rating_df = data.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_rating'}, inplace=True)
avg_rating = data.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating.rename(columns={"Book-Rating": "avg_rating"}, inplace=True)
popular_df = num_rating_df.merge(avg_rating, on='Book-Title')
popular_df = popular_df[popular_df['num_rating'] >= 250]
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author', 'Image-URL-M', 'num_rating', 'avg_rating']]

# Pivot Table for similarity
x = data.groupby('User-ID').count()['Book-Rating'] > 200
active_users = x[x].index
filtered_rating = data[data['User-ID'].isin(active_users)]
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)

# Cosine similarity
similarity_score = cosine_similarity(pt)

# Save all objects
with open('popular.pkl', 'wb') as f:
    pickle.dump(popular_df, f)
with open('pt.pkl', 'wb') as f:
    pickle.dump(pt, f)
with open('books.pkl', 'wb') as f:
    pickle.dump(books, f)
with open('similarity_scores.pkl', 'wb') as f:
    pickle.dump(similarity_score, f)
```
